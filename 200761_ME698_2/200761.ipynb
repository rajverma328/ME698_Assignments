```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x_in = [30, 30, 30, 30, 40, 40, 40, 50, 50, 50, 60, 60, 60, 60]
y = [55.8, 59.1, 54.8, 54.6, 43.1, 42.2, 45.2, 31.6, 30.9, 30.8, 17.5, 20.5, 17.2, 16.9]

x_in = np.array(x_in)
y = np.array(y)

print(x_in)
print(y)
```

```python
# Sample data
x = x_in.reshape((-1, 1))
e=np.empty([14])

# Create and fit the linear regression model
model = LinearRegression()
model.fit(x, y)

# Generate predictions
y_hat = model.predict(x)
w0=model.intercept_
w1=model.coef_
e=y-w0-w1*x_in
# Plot the data points and regression line
plt.scatter(x, y, color='blue', label='Training data')
plt.plot(x, y_hat, color='green', label="Regression line")
plt.xlabel("Age (weeks)",size=20)
plt.ylabel("Shear strength (MPa)",size=20)
plt.title('Influence of propellant age on shear strength',size=15)
plt.legend()
# Set high-definition saving options
dpi = 300
figure = 'regression_propellant2.png'
#lgd = legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5,-0.1))
# Save the figure with high definition
plt.tight_layout()
plt.savefig(figure, dpi=dpi)
plt.show()
print("w1:", w1)
print("w0:", w0)
#print(e)
print(np.sum(e),r2_score(y,y_hat))
```
